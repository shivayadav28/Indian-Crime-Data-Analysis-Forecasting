# 2.Inferential Statistics

Hypothesis testing is a statistical method used to determine whether an observed effect or relationship in a sample of data is likely to represent a real effect or relationship in the population from which the sample was drawn.

This involves stating a null hypothesis, which is the assumption that there is no relationship or effect in the population, and an alternative hypothesis, which is the opposite of the null hypothesis and represents the possibility of a real relationship or effect.

# Z-Test

```python
from statsmodels.stats.weightstats import ztest

crime_data_2012 = crime[crime['YEAR'] == 2012]['TOTAL IPC CRIMES']
crime_data_2023 = crime[crime['YEAR'] == 2013]['TOTAL IPC CRIMES']

z_stat, p_value = ztest(crime_data_2012, crime_data_2023)
print('Z-test results')
print('Z-test statistic:',z_stat)
print('p-value:',p_value)
print()
if p_value < 0.05:
    print("Reject null hypothesis- There is a significant difference in the total IPC crimes between 2012 and 2013.")
else:
    print("Accept null hypothesis- There is no significant difference in the total IPC crimes between 2012 and 2013.")
```

The p-value is greater than 0.05, the null hypothesis is accepted, which means there is no significant difference in the mean total IPC crimes between 2012 and 2013.

# Anova Test

This test is used to determine whether the means of three or more independent groups are significantly different from each other. It is often used in studies that compare the means of multiple groups.

```python
from scipy.stats import f_oneway

states = crime['STATE/UT'].unique()
murder_rates = {}
for state in states:
    murder_rates[state] = crime[crime['STATE/UT'] == state]['MURDER']

# perform the ANOVA test
f_stat, p_value = f_oneway(*murder_rates.values())
print('One-way ANOVA results')
print('F-statistic:', f_stat)
print('p-value:', p_val)
print()
if p_value < 0.05:
    print("Reject null hypothesis- There is a significant difference in the mean number of murders across different states in India.")
else:
    print("Accept null hypothesis- There is no significant difference in the mean number of murders across different states in India.")
```

If the ANOVA test indicates that there is a significant difference in the mean number of murders across different states in India, we can use a post-hoc Tukey test to determine which states have significantly different murder rates. 
If the ANOVA test does not find a significant difference, there is no need to perform a post-hoc test.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
# perform the post-hoc Tukey test
tukey_results = pairwise_tukeyhsd(crime['MURDER'], crime['STATE/UT'])

print(tukey_results.summary())
```

# T-Test

This test is used to determine whether the means of two independent samples are significantly different from each other. It is often used in studies that compare the means of two groups, such as a treatment group and a control group.

```python
from scipy.stats import ttest_ind

state = 'MAHARASHTRA'
state_thefts = crime[crime['STATE/UT']==state]['THEFT']
other_thefts = crime[crime['STATE/UT']!=state]['THEFT']

t_stat, p_val = ttest_ind(state_thefts, other_thefts)

print("State:", state)
print("t-statistic:", t_stat)
print("p-value:", p_val)
print()
if p_val < 0.05:
    print('Reject null hypothesis - Significant difference found between thefts in', state, 'and all other states combined')
else:
    print('Accept null hypothesis - No significant difference found between thefts in', state, 'and all other states combined')
```

The p-value is greater than or equal to 0.05, the null hypothesis is accepted and it is conclude that there is a significant difference between the mean number of thefts in Maharashtra and the mean number of thefts in all other states combined.

# chi-square test

This test is used to determine whether there is a significant association between two categorical variables. It is often used in studies that analyze the frequency of occurrence of different categories or outcomes.

```python
from scipy.stats import chi2_contingency

cont_table = pd.crosstab(crime['STATE/UT'], crime['TOTAL IPC CRIMES'])

# perform chi-square test
chi2, p_val, dof, expected = chi2_contingency(cont_table)

print('Chi-square test results')
print('Chi-square statistic:', chi2)
print('p-value:', p_val)
print('Degrees of freedom:', dof)
print()
if p_val < 0.05:
    print('Reject null hypothesis - There is a significant relationship between STATE/UT and TOTAL IPC CRIMES')
else:
    print('Accept null hypothesis - There is no significant relationship between STATE/UT and TOTAL IPC CRIMES')
```

The p-value is greater than or equal to 0.05, the null hypothesis is accepted and it is concluded that there is no significant relationship between STATE/UT and TOTAL IPC CRIMES.

# Time series analysis

```python
import statsmodels.api as sm
plt.figure(figsize=(15,8))

crime_df = pd.read_csv(r"C:\Users\Admin\Desktop\crime data\crime.csv", parse_dates=['YEAR'], index_col=['YEAR'])

# Resample data to yearly frequency
crime_yearly = crime_df.resample('Y').sum()

# Plot the time series
plt.plot(crime_yearly.index, crime_yearly['TOTAL IPC CRIMES'])
plt.title('Total IPC Crimes over Time')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()

# Decompose the time series into trend, seasonality, and residuals
decomposition = sm.tsa.seasonal_decompose(crime_yearly['TOTAL IPC CRIMES'], model='additive')

# Plot the decomposition results
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 7))
ax1.plot(crime_yearly.index, decomposition.observed)
ax1.set_ylabel('Observed')
ax2.plot(crime_yearly.index, decomposition.trend)
ax2.set_ylabel('Trend')
ax3.plot(crime_yearly.index, decomposition.seasonal)
ax3.set_ylabel('Seasonal')
ax4.plot(crime_yearly.index, decomposition.resid)
ax4.set_ylabel('Residual')
plt.tight_layout()
plt.show()

# Test for stationarity using the Augmented Dickey-Fuller test
adf_result = sm.tsa.stattools.adfuller(crime_yearly['TOTAL IPC CRIMES'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:', adf_result[4])
```

```python
# Plot ACF
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(crime_yearly['TOTAL IPC CRIMES'])
plt.show()
```

```python
# Plot PACF
from statsmodels.graphics.tsaplots import plot_pacf
sm.graphics.tsa.plot_pacf(crime_yearly['TOTAL IPC CRIMES'], lags=5)
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lags')
plt.ylabel('Partial Autocorrelation')
plt.show()
```

```python
# Estimate the autoregressive integrated moving average (ARIMA) model
model = sm.tsa.ARIMA(crime_yearly['TOTAL IPC CRIMES'], order=(1,1,1))
results = model.fit()
print(results.summary())
```

```python
# Plot the actual values and the predicted values
plt.figure(figsize=(15,8))
plt.plot(crime_yearly.index, crime_yearly['TOTAL IPC CRIMES'], label='Actual')
plt.plot(crime_yearly.index, results.predict(start=0, end=len(crime_yearly)-1), label='Predicted')
plt.title('Actual vs Predicted Total IPC Crimes')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```

```python
# Plot the actual values and the forecasted values
plt.figure(figsize=(15,8))
forecast = results.predict(start=len(crime_yearly), end=len(crime_yearly)+4)
plt.plot(crime_yearly.index, crime_yearly['TOTAL IPC CRIMES'], label='Actual')
plt.plot(forecast.index, forecast, label='Forecast')
plt.title('Actual vs Forecasted Total IPC Crimes')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```

```python
def get_crime_level(crime_count):
    if crime_count <= low:
        return 1
    elif crime_count <= medium:
        return 2
    elif crime_count <= high:
        return 3
    else:
        return 4

crime['CRIME_LEVEL'] = crime['TOTAL IPC CRIMES'].apply(get_crime_level)
```
